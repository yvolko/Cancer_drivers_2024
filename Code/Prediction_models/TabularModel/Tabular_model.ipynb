```python
import pandas as pd
```

```python
df = pd.read_csv("../../../Data/Processed_data/Selection_of_regions_for_ML_with_known_oncogenes.csv")
df
```

```python
df_new = df.iloc[:,3:16]
df_new
```

```python
df_new.is_oncogene = df_new.is_oncogene.astype('category')
df_new.dtypes
```

```python
from fastai.text.all import *
from fastai.tabular.all import TabularDataLoaders, tabular_learner, TabularModel
from fastai.metrics import accuracy, Precision, Recall, F1Score
dls = TabularDataLoaders.from_df(df_new, y_names="is_oncogene",
    cont_names = ['rank_expr_norm', 'rank_pubmed_mean_norm', 'rank_tf_norm',
                 'rank_kinase_norm', 'rank_crisp_mean_norm', 'rank_crisp_min_norm',
                 'rank_expr_fold_change_norm', 'rank_is_TSG_norm', 'rank_GO_term_norm',
                 'rank_PPI_norm', 'rank_count_norm','SUM_RANK'])
m = TabularModel(emb_szs = (), n_cont = 12, out_sz=2, ps = [0,0.3,0.2,0], 
                 layers=[400,300,200,100]).eval()
m.layers
```

```python
learn = Learner(dls, m, loss_func=CrossEntropyLossFlat(), metrics=[accuracy, Precision(average='macro'), Recall(average='macro')])
learn.fit_one_cycle(40)
```

```python
df_pred = pd.read_csv('../../../Data/Processed_data/Selection_of_regions_without_known_oncognes.csv')
df_pred_new = df_pred.iloc[:,3:15]
df_pred_new
```

```python
dl = learn.dls.test_dl(df_pred_new) # apply transforms
preds,  _ = learn.get_preds(dl=dl)
preds
```

```python
indices = (preds[:, 1] > 0.9).nonzero(as_tuple=True)[0]
selected_rows = df_pred.iloc[indices.tolist(),2]
selected_rows
```

```python
selected_rows.to_csv('../../../Data/Result_data/Prediction_by_tabular_model.csv', index=True)
```
